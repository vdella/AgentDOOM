```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
log_path = "../logs/ppo_cnn.csv"
df = pd.read_csv(log_path)
df.head()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['step'], df['avg_reward'], label='Avg Reward (100 eps)')
plt.xlabel('Timesteps')
plt.ylabel('Avg Reward')
plt.title('PPO-CNN Tetris Training Curve')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
