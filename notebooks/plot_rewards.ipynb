```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
log_path = "../logs/lr_1e-6_clip_02/ppo_cnn_breakout_500k_iterations.csv"
df = pd.read_csv(log_path)
df.head()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['step'], df['avg_reward'], label='lr=1e-6, clip=0.2')
plt.xlabel('Timesteps')
plt.ylabel('Avg Reward')
plt.title('PPO-CNN Breakout Training Curve for Avg Reward (500k iterations)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
