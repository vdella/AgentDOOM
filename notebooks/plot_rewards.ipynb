```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
plt.figure(figsize=(10, 5))

plt.title('PPO-CNN Breakout Training Curve for Avg Reward (500k iterations)')
plt.xlabel('Timesteps')
plt.ylabel('Avg Reward')

log_path = "../logs/ppo_cnn/breakout/lr_1e-4_clip_01/500k/ppo_cnn.csv"
df = pd.read_csv(log_path)

plt.plot(df['step'], df['avg_reward'], label='lr=1e-4, clip=0.1')

log_path = "../logs/ppo_cnn/breakout/lr_1e-4_clip_02/500k/ppo_cnn.csv"
df = pd.read_csv(log_path)

plt.plot(df['step'], df['avg_reward'], label='lr=1e-4, clip=0.2')

# ======================================================================

log_path = "../logs/ppo_cnn/breakout/lr_1e-5_clip_01/500k/ppo_cnn.csv"
df = pd.read_csv(log_path)

plt.plot(df['step'], df['avg_reward'], label='lr=1e-5, clip=0.1')

log_path = "../logs/ppo_cnn/breakout/lr_1e-5_clip_02/500k/ppo_cnn.csv"
df = pd.read_csv(log_path)

plt.plot(df['step'], df['avg_reward'], label='lr=1e-5, clip=0.2')

# ======================================================================

log_path = "../logs/ppo_cnn/breakout/lr_1e-6_clip_01/500k/ppo_cnn.csv"
df = pd.read_csv(log_path)

plt.plot(df['step'], df['avg_reward'], label='lr=1e-6, clip=0.1')

log_path = "../logs/ppo_cnn/breakout/lr_1e-6_clip_02/500k/ppo_cnn.csv"
df = pd.read_csv(log_path)

plt.plot(df['step'], df['avg_reward'], label='lr=1e-6, clip=0.2')

plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
